```python
# pie plot of diabetes ratio 
plt.figure(figsize=(8,6))
labels = ['No Diabetes', 'Diabetes']
sizes = [Diabetes_data['Diabetes_Status'].value_counts()[1], Diabetes_data['Diabetes_Status'].value_counts()[2]]
colors = ['grey', 'lightcoral']
plt.pie(sizes, labels=labels, autopct='%.1f%%', colors=colors, data = df_vis);

plt.savefig("../Images/pie_chart_diabetes.jpg") #save as jpg

```

```python
#Gender Relationship to Diabetes 
plt.figure(figsize=(20,4))

x= sns.countplot(x='Diabetes_Status',data=df_vis,hue='Sex', palette=["#FFC0CB", "#7BC8F6"])
plt.xticks(rotation=15)
plt.title('Gender Relationship to Diabetes',fontdict={'fontsize':20})
plt.legend(labels = ['Female', 'Male'])
plt.savefig("../Images/gender_v_diabetes.jpg") #save as jpg

plt.show()

```

```python
#Income Correlation to Diabetes
plt.figure(figsize=(20,4))

import pandas as pd
import seaborn as sns

x, y, hue = "Income", "Percentage", "Diabetes_Status"

(df_vis[x]
 .groupby(df_vis[hue])
 .value_counts(normalize=True)
 .rename(y)
 .reset_index()
 .pipe((sns.barplot, "data"), x='Income', y=y, hue='Diabetes_Status'))

plt.savefig("../Images/income_v_diabetes.jpg") #save as jpg
```
